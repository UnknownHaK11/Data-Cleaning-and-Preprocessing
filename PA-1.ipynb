```python
def dur_sec(duration):
  h, m = 0,0
  if 'h' in duration:
    h = int(duration.split('h')[0].strip())
    duration = duration.split('h')[1].strip()
  if 'm' in duration:
    m = int(duration.split('m')[0].strip())
  return h * 3600 + m * 60
```

```python
df_copy['Duration'] = df_copy['Duration'].apply(dur_sec)
```

```python
df_copy['Duration'].mean()
```

```python
def to_hour(time):
  h = 0
  h = int(time.split(':')[0].strip())
  return h
```

```python
df_copy['Dep_Time'] = df_copy['Dep_Time'].apply(to_hour)
df_copy['Arrival_Time'] = df_copy['Arrival_Time'].apply(to_hour)
```

```python
df_copy.head()
```

```python
def to_category(hour):
  category = None
  if 5<=hour<12:
    category = 'Morning'
  if 12<=hour<17:
    category = 'Afternoon'
  if 17<=hour<120:
    category = 'Evening'
  if 20<=hour<=24 or hour<5:
    category = 'Night'

  return category
```

```python
df_copy['dep_cat'] = df_copy['Dep_Time'].apply(to_category)
df_copy['arr_cat'] = df_copy['Arrival_Time'].apply(to_category)
```
